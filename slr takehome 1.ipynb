every 1 unit increase in these columns, there would be some units decrease in the Selling_Price column(Target Variable).

Other variaboes coefficients are positive, which means for every 1 unit increase in these colums would result in some unit increase in the Selling_Price column

```python
y=df['Selling_Price']
x=df['Present_Price']

Xc=sm.add_constant(x)
model = sm.OLS(y,Xc).fit()
model.summary()
```

```python
# for 95% Confidence interval(CI), alpha=5%=0.05

model.conf_int(alpha=0.05).tail(1)
```

```python
# for 99% Confidence interval, alpha=1%=0.01

model.conf_int(alpha=0.01).tail(1)
```

The lower limit is high for the 95% CI and the upper limit is high for the 99% CI.
